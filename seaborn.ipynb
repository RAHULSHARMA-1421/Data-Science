```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=sns.load_dataset("fmri")
df.head()
```

```python
df.shape
```

```python
sns.lineplot(x="timepoint",y="signal",data=df)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=df,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=df,hue="event",style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",style="event",markers=True,data=df)
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.read_csv("fmri.csv")
df.head()
```
